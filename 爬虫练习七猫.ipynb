```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd

# 1. 初始化浏览器
driver_service = Service(executable_path=r"D:\edgedriver_win64\edgedriver_win64\msedgedriver.exe")
edge_options = webdriver.EdgeOptions()
driver = webdriver.Edge(service=driver_service, options=edge_options)

# 2. 初始化存储数据的列表
novel_data = []

# 3. 多页采集的核心逻辑（设置采集页数，或循环到最后一页）
target_url = "https://www.qimao.com/paihang/boy/hot/date"
driver.get(target_url)
time.sleep(5)

# 定义要采集的页数（比如采集3页）
max_page = 3
current_page = 1

while current_page <= max_page:
    print(f"正在采集第 {current_page} 页...")
    
    # ---------- 提取当前页的小说信息 ----------
    novel_items = driver.find_elements(By.XPATH, '//li[@class="rank-list-item"]')
    print(f"第 {current_page} 页共找到 {len(novel_items)} 本小说")
    
    for item in novel_items:
        # 提取标题
        try:
            title = item.find_element(By.XPATH, './/a[@class="s-book-title"]').text.strip()
        except:
            title = "未知标题"
        
        # 提取作者
        try:
            author = item.find_element(By.XPATH, './/a[contains(@href, "zuozhe")]').text.strip()
        except:
            author = "未知作者"
        
        # 提取简介
        try:
            intro = item.find_element(By.XPATH, './/span[@class="s-book-intro"]').text.strip()
        except:
            intro = "暂无简介"
        
        # 提取热度
        try:
            hot_score = item.find_element(By.XPATH, './/span[@class="rank-change-num"]').text.strip()
        except:
            hot_score = "0"
        
        # 存入数据
        novel_data.append({
            "小说标题": title,
            "作者": author,
            "简介": intro,
            "热度": hot_score
        })
    
    # ---------- 切换到下一页 ----------
    if current_page < max_page:
        try:
            # 点击“下一页”按钮（需替换为实际的分页按钮XPath）
            next_page_btn = driver.find_element(By.XPATH, '//div[@class="qm-page-btn"]')
            next_page_btn.click()
            time.sleep(5)  # 等待页面加载
            current_page += 1
        except:
            print("已到最后一页，停止采集")
            break
    else:
        break

# 4. 关闭浏览器+保存数据
driver.quit()
```

```python
# 保存数据
df = pd.DataFrame(novel_data)
df.to_csv("data/七猫小说数据.csv", index=False, encoding='utf-8-sig')

# 关闭浏览器
driver.quit()
print("爬取完成！数据已保存到当前目录的「data/七猫小说数据.csv」")
```
